# Train, validation, and test dataset

```python
# The df.sample(frac=1) suffles the entire dataset
# We have to use the to_numpy() method to 
# [int(0.6*len(df))] tells the split function where to split the datasets
train, valid, test = np.split(df.sample(frac=1), [int(0.6*len(df)), int(0.8*len(df))])
```

```python
#We need to scale the dataset because some of the values in the columns are way bigger compared to others
def scale_dataset(df, oversample=False):
    # Remember -> values brings up the values in the series
    x = df[df.columns[:-1]].values
    y = df[df.columns[-1]].values

    scaler = StandardScaler()
    #Take the values in x, run it through the standard transformer, assign it to x
    x = scaler.fit_transform(x)

    if oversample:
        ros = RandomOverSampler()
        x,y = ros.fit_resample(x,y)

    # We can call hstack, but we have a problem
    # x is a 2d matrix, but y is a single column
    # (-1,1) is telling the computer to infer the number of rows
    data = np.hstack((x, np.reshape(y, (-1,1))))

    return data, x, y
```

```python
print(len(train[train['class']==1])) #gamma 
print(len(train[train['class']==0])) #hadrons
```

```python
train, train_x, train_y = scale_dataset(train, True)

# We are not oversampling for the validation and test dataset because we want the model to face uneven datasets
valid, valid_x, valid_y = scale_dataset(valid, False)
test, test_x, test_y = scale_dataset(test, False)
```

# Naive Bayes
Probability of something happening given B.
P(A|B) = $\frac{P(B|A)*P(A)}{P(B)}$

Imagine we have a couple of categories, the categories are cats, dogs, lizards, zebras, etc. Then we have a new data point (feature vector). Naive bayes is trying to find the probability this new feature vector fits into the categories. The category with the highest score will be where this new feature vector belongs. 

- We want to create a filter to remove spam emails
- We take all of the words from normal emails and create a histogram, we do the same for the words in the spam emails.
- We then determine the probability of all those words occuring
- Then we take an initial guess (taken from the training data)
- We have an email with dear(0.47) friend(0.29)
- We do this math -> P(N) * P(Dear|N) * P(Friend|N) = 0.09
- Same math for spam email P(S) * 0.29 * 0.14 = 0.01

- Doing this method is simple, but can lead to a lot of issues.
- If we have a word that does not show up in the spam/normal folder, the probabilty of that word showing up is 0. When that word shows up, the probabilty will automatically become zero, causing incorrect classifications.
- To avoid this, we add 1 to all of the probabilities. (This does not change the total numbers of spam/normal messages in the dataset

- Why is Naive Bayes naive?
- The 'total probability' we determine does not take work order into accout. So, if we had an email with Dear Friend, and an email with Friend Dear, both would be classified with the same score. This causes incorrect classifications.
- Naive Bayes has high bias, and low variance. 

```python
from sklearn.naive_bayes import GaussianNB
```

- We could also calculate the p-value to determine if R^2 is statistically significant
- We can use the line to predict size given weight.
- Using data to predict something is a form of machine learning

## Logistic Regression
- Predicts if something is true or false instead of their exact values
- There is no line in logistic regression, it is an S shaped curve.
- Curve tells us the probabilty a mouse is obese based on its weight
- Heavy mouse -> high chance the mouse is obese
- Medium weight mouse -> 50% chance the mouse is obese
- If the probability of the mouse being overweight is >50%, the model will classify it as obese, <50%, the model will classify it as not obese.

- LSRL -> used to find lin-reg line, we also have R^2 (how much of the variance in the dependent variable is explained by the independent variable
- Logistic regression is all probabilites, so there is no concept of R^2.
- Instead, we use maximum likelihood

# Neural Network

- Neural networks are inspired by the brain
- Nueron is a thing that holds a number (0-1)
- We start with a 28*28 grid of neurons. Each of these neurons holds a grayscale value (0.0 for black, 1.0 for white)
- Each of these grayscale values is an activation function, think of it like turning on the nueron!
- All 784 of these neurons make up the first layer of our network
- The last layer of our network has 10 neurons, these neurons have an activation function of how much a image looks like the corresponding number (0-9)
- In between, are the hidden layers, 16 nuerons each
- Activation in one layer, corresponds to activation in another layer
- Similiar to how human brains work. When some neurons fire, it causes other ones to become activated as well.
- On each layer, a certain combination of nuerons are activated, corresponding to if certain pieces of a number are identified
- Certain combinations lead to corresponding numbers
- But how in the world does the nueron recognize pieces of a 9? Like how does it recognize a loop?
- In the pervious layer, the nuerons activate when edges are detected
## How do we know if there is an edge in the picture?
- Every neuron from the first layer has a connection to 1 neuron in the next later. Each of these connections are given a weight
- To get the value of the new neuron, take the activation values from the first layer, mutliply it by the weight, and add them all up
- But this value of the new neuron can be anything, right?
- Yes, this value can be anything, but we use the sigmoid function to compress down the answer to 0 to 1. $\sigma (x) = \frac{1}{1 + e^{-x}}$
- Value close to 1 signifies that a corner was detected, a value close to 0 signifies that a corner was not detected. 
- Activation is how positive the weighted sum is.
- We can also specify certain parameters -> only light up when the value of the weighted sum is > 10. This is called the bias
- The weights tells us what the pixel pattern is
- The bias tells the neuron when to actually light up
- Learning -> finding the right weight and biases to solve the problem.
- This equation is hard to know, lets use matrices
- All the weights from the connections are in one martix, and all of the activation values are stored in a matrix with one column.
- After the matrix is multiplied, we add a vector of biases.  v

# How we know what our weighted values are?
- We set all the weights to 0, except for the region of interest
- If there are "activated neurons in that region of interest, then we give it a positive weight
- We can also set a negative region, where if there are activated pixels in that region, there will be a negative weight.
